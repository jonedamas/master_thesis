```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings("ignore")
```

```python
ticker = 'SPY'

data = yf.download(ticker, start='2010-01-01')[['Adj Close']]

data['log_ret'] = np.log(data['Adj Close']/data['Adj Close'].shift(1))

realized_vol = data['log_ret'].groupby(pd.Grouper(freq='M')).std()

realized_vol = realized_vol.dropna()

realized_vol.plot()
```
